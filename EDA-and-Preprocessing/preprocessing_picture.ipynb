```python
import os
from pdf2image import convert_from_path
import pytesseract
import cv2
import numpy as np
```

```python
def preprocess(img):
    img = np.array(img)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # tốt hơn threshold fix
    thresh = cv2.threshold(
        gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
    )[1]
    
    return thresh


def extract_text_from_pdf(pdf_path):
    try:
        images = convert_from_path(pdf_path)
    except Exception as e:
        raise RuntimeError(f"Lỗi convert PDF: {pdf_path}") from e

    full_text = ""
    
    for i, img in enumerate(images):
        try:
            processed = preprocess(img)
            text = pytesseract.image_to_string(processed, lang='vie')
            full_text += text + "\n"
        except Exception as e:
            raise RuntimeError(f"Lỗi OCR trang {i} trong {pdf_path}") from e
    
    return full_text


def process_folder(pdf_folder, output_folder):
    os.makedirs(output_folder, exist_ok=True)
    
    for file in os.listdir(pdf_folder):
        if file.endswith(".pdf"):
            pdf_path = os.path.join(pdf_folder, file)
            print(f"Processing: {file}")
            
            try:
                text = extract_text_from_pdf(pdf_path)
                
                output_path = os.path.join(
                    output_folder, file.replace(".pdf", ".txt")
                )
                with open(output_path, "w", encoding="utf-8") as f:
                    f.write(text)
                    
            except Exception as e:
                # raise luôn (theo yêu cầu của bạn)
                raise RuntimeError(f"Lỗi file {file}") from e

NEU_docs_path = "data/raw/NEU_docs"
GDDT_docs_path = "data/raw/GDDT_docs"
```
